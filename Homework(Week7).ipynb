#### Given a set of blue points, calculate the parameters of the fitted quadratic curve using the gradient descent method.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = 1.5*x*x + 0.2 + 0.2*np.random.randn(100)
plt.scatter(x, y)
plt.plot(x, 1.5*x*x + 0.2, 'r-', lw=3)
plt.show()
```

```python
def step_gradient_quad(b_current, w_current, points, learningRate):
    b_gradient = 0
    w_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((w_current * x**2) + b_current))
        w_gradient += -(2/N) * x**2 * (y - ((w_current * x**2) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_w = w_current - (learningRate * w_gradient)
    return [new_b, new_w]

# 
learning_rate = 0.1
b = 0
w = 0
num_iterations = 1001
points = np.zeros((100, 2))
points[:,0] = x
points[:,1] = y

# 
for i in range(num_iterations):
    if i % 100 == 0:
        print('iteration %d, error is %.4f' % (i, np.mean((y - (w * x**2 + b)) ** 2)))
        
        plt.scatter(x, y)
        plt.plot(x, w*x**2 + b , 'r-', lw=2)
        plt.show()
        
    b, w = step_gradient_quad(b, w, points, learning_rate)

print("b = %.4f, w = %.4f" % (b, w))
```
